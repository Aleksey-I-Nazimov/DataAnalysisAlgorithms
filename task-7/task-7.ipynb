<b>По заданиям 1 и 3</b> (в особенности по 3 части). Ниже код, который использует циклы при расчете минимально, недопуская вложенности for-for. Сложность обучения такого KNN - O(N). Сложность предсказания O (N * M), где N - это количество векторов а М - это количество признаков или координат в каждом векторе. Это конечно не константная работа, но все же лучше чем О(N^2), как это было в теоретической части. Константное вычисление может быть достигнуто засчет усложнения стадии обучения. По сути, там нужно будет выстраивать M-мерное пространство и общитывать каждый его сегмент, чтобы при распознавании константно по сложности находить нужную ячейку и на основе ее внутренних значений забирать прогноз. Такая модель будет сложна для программирования и прожорлива по ресурсам компьютера на стадии обучения алгоритма. Простой и довольно производительный алгоритм показан ниже.

```python

class Row:
    def __init__(self, x_row, y_label):
        self.x_row = x_row;
        self.y_label = y_label;
        
    def get_value (self,column_number):
        return self.x_row[column_number];
    
    def calculate_distance (self, x):
        return np.linalg.norm(self.x_row-x);
    
    def get_label(self):
        return self.y_label;
    
    def __str__(self):
        return "Row={} l={}".format(self.x_row,self.y_label)
    
    def __repr__(self):
        return "Row={} l={}".format(self.x_row,self.y_label)
    
    def __eq__(self,other):
        if self is other:
            return True;
        if np.array_equal(self.x_row,other.x_row) and np.array_equal(self.y_label,other.y_label):
            return True;
        else:
            return False;
    def __hash__(self):
        return hash((str(self.x_row), str(self.y_label)));

    
def make_row_group_map (rows,column_number):
    rows = sorted(rows, key=lambda r: r.get_value(column_number))
    row_map = {};
    for row in rows:
        value = row.get_value(column_number);
        if value not in row_map: row_map[value]=[];
        row_map[value].append(row)
    return row_map;


def get_sorted_keys (row_map):
    keys=[];
    for key in row_map: keys.append(key);
    keys = sorted(keys);
    return keys;


def search_index_rule (root_index,limit):
    neighbour_indexes=[];
    if root_index==0:
        neighbour_indexes.append(root_index);
        neighbour_indexes.append(root_index+1);
        #print ("Selected from the left: {}".format(suitable_rows));
    elif root_index==limit:
        neighbour_indexes.append(root_index-1);
        #print ("Selected from upper right: {}".format(suitable_rows));
    elif root_index==limit-1:
        neighbour_indexes.append(root_index);
        neighbour_indexes.append(root_index-1);
        #print ("Selected from the right: {}".format(suitable_rows));
    else:
        neighbour_indexes.append(root_index+1);
        neighbour_indexes.append(root_index);
        neighbour_indexes.append(root_index-1);
        #print ("Selected from internal: {}".format(suitable_rows));
    return neighbour_indexes;
    
    
class Column:
    
    def __init__(self,rows,number):
        self.row_map = make_row_group_map(rows,number);
        self.number = number;
        
    def search_suitable (self, row_value):
        
        unique_row_values = get_sorted_keys(self.row_map);
        root_index = np.searchsorted(np.array(unique_row_values),row_value);
        neighbour_indexes = search_index_rule(root_index=root_index, limit=len(unique_row_values));
        
        suitable_rows=[];
        for neighbour_index in neighbour_indexes:
            rows=self.row_map[unique_row_values[neighbour_index]];
            suitable_rows.extend(rows);
        
        return suitable_rows;
    
    def __str__(self):
        return "Column: n={}, row_map={},".format(self.number,self.row_map);

    
class Matrix:
    def __init__(self,columns):
        self.columns = columns;
    
    def search_suitable (self,x,neighbours_number):
        
        suitable_row_list=[];
        for index in range(0,len(self.columns)):    
            suitable_row_list.extend(self.columns[index].search_suitable(row_value=x[index]));
            
        suitable_row_set=set(suitable_row_list);
        suitable_row_list = list (sorted (suitable_row_set,key=lambda r: r.calculate_distance(x)));
        suitable_row_list = suitable_row_list[0:neighbours_number];

        # Just small debug:
        #if x[0]>8.5 and x[0]<8.7 and x[1]>4.1 and x[1]<4.2:
        #    print("x={} list={} set{} ".format(x,suitable_row_list,suitable_row_set))

        return suitable_row_list;
    
    def __str__(self):
        xtr="";
        for column in self.columns:
            xtr += str(column);
        return "Matrix: {}".format(xtr);

    
class MatrixFactory:
        
    def make (self,x,y):
        rows = [];
        for x_row,y_label in zip(x,y): rows.append(Row(x_row,y_label));
        
        Qv,Qp = x.shape;
        columns = [];
        for j in range(0,Qp): columns.append(Column(rows,j));
        matrix = Matrix(columns);
        return matrix;

    
def make_class_map (rows):
    class_map = {};
    for row in rows:
        label = row.get_label();
        if label not in class_map: class_map[label] = 0;
        class_map[label] = class_map[label]+1
    return class_map;
    
    
class WeightedResults:
    
     def calculate_results(self,rows,x):
        r=[];
        for row in rows: r.append(row.calculate_distance(x));
        r_max = np.amax(r);
        r_min = np.amin(r);
        r_min = r_min==0.0 if 1.0/len(rows) else r_min;
        cm = make_class_map(rows);
        
        ref_w=0;ref_label=0;
        for row in rows:
            w = cm[row.get_label()]*r_max/(r_min+row.calculate_distance(x));
            if ref_w<w:
                ref_w=w;
                ref_label=row.get_label();
        return ref_label;

    
class CountedResults:
    def calculate_results(self,rows,x):
        class_map = make_class_map(rows);
        selected_label=0;cnt=-1;
        for label in class_map:
            if cnt<class_map[label]:
                selected_label = label;
                cnt=class_map[label];
        return selected_label;

    
class Knn:
    def __init__(self,result_counter=CountedResults()):
        self.matrix_factory = MatrixFactory();
        self.matrix = None;
        self.number = None;
        self.result_counter = result_counter;
    
    def fit (self,X,Y,neighbours_number):
        self.matrix = self.matrix_factory.make(X,Y);
        self.neighbours_number = neighbours_number;
        return self;
    
    def predict (self,X):
        y = [];
        for x_row in X: y.append(self.predict_vector(x_vector=x_row));
        return np.array(y);
    
    def predict_vector (self,x_vector):
        suitable_rows = self.matrix.search_suitable(x=x_vector,neighbours_number=self.neighbours_number);
        return self.result_counter.calculate_results(rows=suitable_rows,x=x_vector);
    
```

```python
X, Y = load_iris(return_X_y=True)

#print (X.shape)
#X = np.stack((X[:, 0], X[:, 1], Y),axis=1);
#print (X)
#print (Y.shape)

X = np.stack((X[:, 0], X[:, 1]),axis=1);
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.2,
                                                    random_state=1,
                                                    stratify=Y);
cmap = ListedColormap(['red', 'green', 'blue'])
plt.scatter(X[:,0], X[:,1], c=Y, cmap=cmap);
```

```python
# The method of making grid arrays with constant step and border
def make_meshgrid(data, step=.05, border=0.3):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))


def visualize_scatter(predictor,train_data, train_labels, scatter_colors, background_colors, title):
    
    xx, yy = make_meshgrid(train_data)
    mesh_predictions = np.array(predictor.predict(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
    plt.pcolormesh(xx, yy, mesh_predictions, cmap = background_colors)
    plt.scatter(train_data[:, 0], train_data[:, 1], c = train_labels, cmap = scatter_colors)
    plt.title(title)

def accuracy(pred, y):
    return (sum(pred == y) / len(y))
```

Рассмотрим влияние окружения без взвешивания
